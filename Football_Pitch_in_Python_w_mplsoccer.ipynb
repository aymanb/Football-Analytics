```python
from mplsoccer import Pitch, VerticalPitch
pitch = Pitch()
fig, ax = pitch.draw()
```

```python
pitch_ver = VerticalPitch()
fig, ax = pitch_ver.draw()
```

```python
pitch = Pitch()
fig, ax = pitch.grid(nrows = 2, ncols = 2)
```

```python
pitch = Pitch(half=True)
fig, ax = pitch.draw()
```

```python
pitch = VerticalPitch(half=True)
fig, ax = pitch.draw()
```

```python
pitch = VerticalPitch(pitch_color= 'grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
```

```python
pitch = Pitch(pitch_color= 'grass', line_color='white', stripe=True)
fig, ax = pitch.draw()
```
